```python
df = sf.load_balance(variant='annual', market='us')   #us quarterly balance sheet 
```

```python
df.iloc[[0,11],:]
```

```python
df.shape # 14,623 rows, 27 columns, 
df['SimFinId'].nunique() # 1910 company 
df.info()
```

```python
num_cols = ['Fiscal Year','Shares (Basic)',
            'Shares (Diluted)',
            'Cash, Cash Equivalents & Short Term Investments',
            'Accounts & Notes Receivable',
            'Inventories',
            'Total Current Assets',
            'Property, Plant & Equipment, Net',
            'Long Term Investments & Receivables',
            'Other Long Term Assets',
            'Total Noncurrent Assets',
            'Total Assets',
            'Payables & Accruals',
            'Short Term Debt',
            'Total Current Liabilities',
            'Long Term Debt',
            'Total Noncurrent Liabilities',
            'Total Liabilities',
            'Share Capital & Additional Paid-In Capital',
            'Treasury Stock',
            'Retained Earnings',
            'Total Equity',
            'Total Liabilities & Equity'
           ]   
```

```python
df.columns[df.isnull().any()].tolist() # 18 columns with null value 
```

```python
df['Shares (Basic)'].fillna(-1000000000, inplace = True)
df['Shares (Diluted)'].fillna(-1000000000, inplace = True)
df['Cash, Cash Equivalents & Short Term Investments'].fillna(-1000000000, inplace = True)
df['Accounts & Notes Receivable'].fillna(-1000000000, inplace = True)
df['Inventories'].fillna(-1000000000, inplace = True)
df['Total Current Assets'].fillna(-1000000000, inplace = True)
df['Property, Plant & Equipment, Net'].fillna(-1000000000, inplace = True)
df['Long Term Investments & Receivables'].fillna(-1000000000, inplace = True)
df['Other Long Term Assets'].fillna(-1000000000, inplace = True)
df['Total Noncurrent Assets'].fillna(-1000000000, inplace = True)
df['Payables & Accruals'].fillna(-1000000000, inplace = True)
df['Short Term Debt'].fillna(-1000000000, inplace = True)
df['Long Term Debt'].fillna(-1000000000, inplace = True)
df['Total Noncurrent Liabilities'].fillna(-1000000000, inplace = True)
df['Share Capital & Additional Paid-In Capital'].fillna(-1000000000, inplace = True)
df['Treasury Stock'].fillna(-1000000000, inplace = True)
df['Retained Earnings'].fillna(-1000000000, inplace = True)
df['Total Equity'].fillna(-1000000000, inplace = True)
```

```python
for i, col in enumerate(df.columns):
    if col in num_cols:
        plt.figure(i)
        sns.distplot(df[col])
```
